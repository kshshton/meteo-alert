```python
import pandas as pd

from utils import plot, get_week_metadata
```

```python
week_metadata = get_week_metadata()
```

```python
def plot_weekly_cloud_cover_trend(observation_hour_range = 3, cloudiness_threshold = 80):
    for day_metadata in week_metadata:
        date = day_metadata['date']
        day = [{'time': hour['time'], 'cloud': hour['cloud']} for hour in day_metadata['hour']]
        df = pd.DataFrame(day)
        df['time'] = pd.to_datetime(df['time'])
        df['rolling_cloudiness'] = df['cloud'].rolling(window=observation_hour_range, min_periods=1).mean()
        df['above_threshold'] = df['rolling_cloudiness'] > cloudiness_threshold
        plot(df, ['cloud'], ['Cloudiness'], 'Cloudiness (%)', f'Cloud Cover Trend ({date})')
```

```python
observation_hour_range = 3
custom_cloudiness_threshold = 80

plot_weekly_cloud_cover_trend(observation_hour_range, custom_cloudiness_threshold)
```
