```python
import pandas as pd

from utils import plot, get_week_metadata
```

```python
week_metadata = get_week_metadata()
```

```python
def plot_weekly_humidity_cover_trend(observation_hour_range = 3, humidity_min_threshold = 30, humidity_max_threshold = 80):
    for day_metadata in week_metadata:
        date = day_metadata['date']
        day = [{'time': hour['time'], 'humidity': hour['humidity']} for hour in day_metadata['hour']]
        df = pd.DataFrame(day)
        df['time'] = pd.to_datetime(df['time'])
        df['rolling_humidity'] = df['humidity'].rolling(window=observation_hour_range, min_periods=3).mean()
        df['above_threshold'] = (df['rolling_humidity'] < humidity_min_threshold) | (df['rolling_humidity'] > humidity_max_threshold)
        plot(df, ['humidity'], ['Humidity'], 'Humidity (%)', f'Humidity Cover Trend ({date})')
```

```python
custom_observation_hour_range = 3
custom_humidity_min_threshold = 30
custom_humidity_max_threshold = 80

plot_weekly_humidity_cover_trend(custom_observation_hour_range, custom_humidity_min_threshold, custom_humidity_max_threshold)
```
