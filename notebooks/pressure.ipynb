```python
import pandas as pd

from utils import plot, get_week_metadata
```

```python
week_metadata = get_week_metadata()
```

```python
def plot_weekly_pressure_trend(observation_hour_range = 3, significant_diff_threshold = 5, min_pressure_threshold = 1005):
    for day_metadata in week_metadata:
        date = day_metadata['date']
        day = [{'time': hour['time'], 'pressure_mb': hour['pressure_mb']} for hour in day_metadata['hour']]
        df = pd.DataFrame(day)
        df['time'] = pd.to_datetime(df['time'])
        df['above_threshold'] = (df['pressure_mb'].diff(periods=observation_hour_range) >= significant_diff_threshold) | (df['pressure_mb'] < min_pressure_threshold)
        plot(df, ['pressure_mb'], ['Pressure'], 'Pressure (hPa)', f'Atmospheric Pressure Trend ({date})')
```

```python
custom_observation_hour_range = 4
custom_significant_diff_threshold = 3

plot_weekly_pressure_trend(custom_observation_hour_range, custom_significant_diff_threshold)
```
