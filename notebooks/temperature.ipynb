```python
import pandas as pd

from core import plot, WEEK_METADATA
```

```python
def plot_weekly_temperature_trend(significant_diff_hour_range=3, significant_diff_threshold=6):
    for day_metadata in WEEK_METADATA:
        date = day_metadata['date']
        day_temp = [{'time': hour['time'], 'temp_c': hour['temp_c'], 'feelslike_c': hour['feelslike_c']} for hour in day_metadata['hour']]
        df = pd.DataFrame(day_temp)
        df['time'] = pd.to_datetime(df['time'])
        df['temp_change'] = df['temp_c'].diff(periods=significant_diff_hour_range)
        df['feelslike_change'] = df['feelslike_c'].diff(periods=significant_diff_hour_range)
        df['above_threshold'] = (df['temp_change'].abs() >= significant_diff_threshold) | (df['feelslike_change'].abs() >= significant_diff_threshold)
        plot(df, ['temp_c', 'feelslike_c'], ['Official', 'Feels-like'], 'Temp (Â°C)', f'Official vs. Feels-like Temperature Trend ({date})')
```

```python
custom_significant_diff_hour_range = 3
custom_significant_diff_threshold = 6

plot_weekly_temperature_trend(custom_significant_diff_hour_range, custom_significant_diff_threshold)
```
